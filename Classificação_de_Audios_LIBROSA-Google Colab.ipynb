# Criação do dataframe
- E divisão treino-teste para validação cruzada

````python
df_sound = []
for cada_caminho in sound_list:
    # cada_caminho vem no formato: data\\50\\3_50_38.wav
    classe = int(cada_caminho.split('/')[-1][0]) # o [-1] representa o último elemento enquanto o [0] representa o primeiro dígito
    df_sound.append([cada_caminho, classe])

df_sound = pd.DataFrame(df_sound, columns=['path', 'target'])

df_train, df_test = train_test_split(df_sound, test_size=0.1, shuffle=True, stratify=df_sound['target'])

print (f'df_train.shape: {df_train.shape}, df_test.shape: {df_test.shape}')
````

# Função que extrai as features dos audios

````python
def extrair_features(sound_path):
    data_array, sample_rate = librosa.load(sound_path, res_type='kaiser_fast')
    mfcc = librosa.feature.mfcc(y=data_array, sr=sample_rate, n_mfcc = 40)

    # Pegando as médias para o eixo 0
    return mfcc.mean(axis=1).reshape(1, -1) / 100

sound_features = extrair_features(choice(sound_list))
````

# Criação da classe dataset

````python
class custom_dataset(Dataset):
    
    def __init__(self, df_):
        self.df_ = df_

        vetor_final, classes = [], []
        for caminho_cada_som, classe in tqdm(self.df_.values, ncols=70):
            vetor_final.append(extrair_features(caminho_cada_som))
            classes.append(classe)
        
        self.x_tensor = torch.tensor(vetor_final)
        self.y_tensor = torch.tensor(classes)
    
    def __len__(self):
        return len(self.df_)
    
    def __getitem__(self, index):
        return self.x_tensor[index], self.y_tensor[index]
````

# Instanciando os datasets
- Além disso, serão criados seus respectivos dataloaders

````python
train_dataset = custom_dataset(df_train)
test_dataset = custom_dataset(df_test)
BATCH_SIZE = 16
train_dataloader = DataLoader(train_dataset, batch_size = BATCH_SIZE, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size = BATCH_SIZE, shuffle=True)
````

# Criação da classe modelo

````python
class custom_model(nn.Module):
    def __init__(self, input_example, n_classes):
        super(custom_model, self).__init__()

        self.n_classes = n_classes
        backbone = nn.Sequential(
            nn.Conv1d(in_channels=1, out_channels=16, kernel_size=5, stride=1),
            nn.LeakyReLU(),
            nn.Conv1d(in_channels=16, out_channels=16, kernel_size=5, stride=2),
            nn.LeakyReLU(),
            nn.Dropout(),
            nn.Flatten()
        )

        output = backbone(input_example)
        input_features_linear = output.shape[-1]

        self.net = nn.Sequential(
            backbone,
            nn.Linear(in_features=input_features_linear, out_features=256),
            nn.LeakyReLU(),
            nn.Dropout(),
            nn.Linear(in_features=256, out_features=128),
            nn.LeakyReLU(),
            nn.Linear(in_features=128, out_features=self.n_classes)
        )
    
    def forward(self, x):
        return self.net(x)

x_tensor, y_tensor = next(iter(train_dataloader))
model = custom_model(x_tensor, 10)
print (f'x_tensor.shape: {x_tensor.shape}')
output = model(x_tensor)
print (f'output.shape: {output.shape}')
````

- Instanciando modelo e otimizador e DEVICE

````python
DEVICE = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print (f'device: {DEVICE}')
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

model.to(DEVICE)
loss_fn = loss_fn.to(DEVICE)
````

# Função Validação

````python
def validar(model_, loss_fn_, dataloader_, device_):
    model_.eval()
    with torch.no_grad():
        test_loss, total_itens = 0, 0
        ypreds, ytrues = [], []
        for x, y in dataloader_:
            x, y = x.to(device_), y.to(device_)
            output = model_(x)
            loss = loss_fn_(output, y)
            test_loss += loss.item()
            total_itens += len(x)

            probs_max, indices_max = output.max(dim=1)
            ypreds.append(indices_max.detach().cpu().numpy())
            ytrues.append(y.detach().cpu().numpy())
    
    ypreds = np.hstack(ypreds)
    ytrues = np.hstack(ytrues)

    test_loss = test_loss/total_itens
    accuracy_test = accuracy_score(ytrues, ypreds)

    return test_loss, accuracy_test
````

# Treinamento

````python
cabecalho = ['epoch', 'train_loss', 'test_loss', 'accuracy_test']
print ('|'.join([str(c).center(25) for c in cabecalho]))

resultados = []
for epoch in range(40):

    model.train()
    train_loss, total_itens = 0, 0
    for x, y in train_dataloader:
        x, y = x.to(DEVICE), y.to(DEVICE)
        
        optimizer.zero_grad()
        output = model(x)
        loss = loss_fn(output, y)
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()
        total_itens += len(x)
    
    test_loss, accuracy_test = validar(model, loss_fn, test_dataloader, DEVICE)
    train_loss = train_loss/total_itens
    vetor = [epoch, train_loss, test_loss, accuracy_test]
    resultados.append(vetor)
    print ('|'.join([str(round(v, 3)).center(25) for v in vetor]))
````

# Análise de resultados

````python
resultados = np.array(resultados)
plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.plot(resultados[:,0], resultados[:,1], label='train_loss')
plt.plot(resultados[:,0], resultados[:,2], label='test_loss')
plt.title('Losses')

plt.subplot(1, 2, 2)
plt.plot(resultados[:,0], resultados[:, 3])
plt.title('Accuracy test')

plt.show()
````

# Matriz de confusão das classes preditas

````python
ytrues, ypreds = [], []
model.eval()
with torch.no_grad():
    for x, ytrue in tqdm(test_dataloader):
        x, y = x.to(DEVICE), y.to(DEVICE)
        ypred = model(x)

        ytrues.append(ytrue.detach().cpu().numpy())
        max_probs, indices_max = ypred.max(dim=1)
        ypreds.append(indices_max.detach().cpu().numpy())
    
ypreds = np.hstack(ypreds)
ytrues = np.hstack(ytrues)

plt.figure(figsize=(16, 16))
plt.title('Matriz de Confusão')
sns.heatmap(confusion_matrix(ytrues, ypreds), annot=True, cbar=False, fmt='0', cmap='YlGnBu')
plt.xlabel('Valores preditos')
plt.ylabel('Valores reais')
plt.show()
````

# Relatório das predições

````python
print (classification_report(ytrues, ypreds))
````
